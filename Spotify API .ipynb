# Completed Work

```python
# 완성본

# Audio Features
dance = []
energy = []
key = []
loudness = []
speech = []
acoustic = []
instrument = []
live = []
valence = []
tempo = []
duration = []
time = []


#Track Features
t_name = []
album_name = []
t_pop = []

#Artist Features
a_name = []
a_pop = []
a_genre = []
a_follower = []

#id
u_id = []

permalink = ["https://open.spotify.com/playlist/37i9dQZEVXbNG2KDcFcKOF?si=1333723a6eff4b7f",
            "https://open.spotify.com/playlist/6fyW9hQflhUC4zSOFRCrXt", "https://open.spotify.com/playlist/0X0wx8BxtzTdw7sNRjJT6j",
            "https://open.spotify.com/playlist/0X0wx8BxtzTdw7sNRjJT6j", "https://open.spotify.com/playlist/69Ao4zlaC1AAJDBDC0ETDO",
            "https://open.spotify.com/playlist/37i9dQZF1DX9vFCryQCDjh", "https://open.spotify.com/playlist/61iOYD59VWPF0AQpqJaPot",
            "https://open.spotify.com/playlist/79DI01JNxFtqJoUq3Okc83", "https://open.spotify.com/playlist/5u43GUXUOq8TPFMhRvApWB", 
            "https://open.spotify.com/playlist/37i9dQZF1DX4winw66sZTE","https://open.spotify.com/playlist/37i9dQZF1DX5fMwWEdzkDc",
            "https://open.spotify.com/playlist/37i9dQZF1DX8j2fTnASZ3f","https://open.spotify.com/playlist/37i9dQZF1DXe4qmDjDW0Ug",
            "https://open.spotify.com/playlist/37i9dQZF1DXdTb8AG95jne","https://open.spotify.com/playlist/37i9dQZF1DX1wdZM1FEz79",
            "https://open.spotify.com/playlist/37i9dQZF1DX3Cii9ioWjQP"]



for link in permalink:
    link = link.split("/")[-1].split("?")[0]
    for track in tqdm(sp.playlist_tracks(link)["items"]):
        try:
            track_uri = track["track"]["uri"]
            features = sp.audio_features(track_uri)[0]
            d = features["danceability"]
            e = features["energy"]
            k = features["key"]
            l = features["loudness"]
            s = features["speechiness"]
            a = features["acousticness"]
            i = features["instrumentalness"]
            li = features["liveness"]
            v = features["valence"]
            t = features["tempo"]
            d = features["duration_ms"]
            tm = features["time_signature"]

            dance.append(d)
            energy.append(e)
            key.append(k)
            loudness.append(l)
            speech.append(s)
            acoustic.append(a)
            instrument.append(i)
            live.append(li)
            valence.append(v)
            tempo.append(t)
            duration.append(d)
            time.append(tm)

            #ID
            user_id = track["track"]["id"]
            u_id.append(user_id)
            
            #Track name
            track_name = track["track"]["name"]
            t_name.append(track_name)

            #Album
            album = track["track"]["album"]["name"]
            album_name.append(album)

            #Popularity of the track
            track_pop = track["track"]["popularity"]
            t_pop.append(track_pop)

            #Main Artist
            artist_uri = track["track"]["artists"][0]["uri"]
            artist_info = sp.artist(artist_uri)


            #Name, popularity, genre, followers
            artist_name = track["track"]["artists"][0]["name"]
            a_name.append(artist_name)

            artist_pop = artist_info["popularity"]
            a_pop.append(artist_pop)

            artist_genres = artist_info["genres"]
            a_genre.append(artist_genres)

            artist_followers = artist_info["followers"]["total"]
            a_follower.append(artist_followers)
        except:
            pass


    spotify_df = pd.DataFrame({'id':u_id, 'artist_name' : a_name, 'track_name' : t_name, 'album_name' : album_name,
                             'artist_genre' : a_genre, 'artist_popularity' : a_pop, 'track_popularity' : t_pop,
                             'artist_followers' : a_follower, 'danceability' : dance, 'energy' : energy,
                            'key' : key, 'loudness' : loudness, 'speechiness' : speech, 'acousticness' : acoustic,
                            'instrumentalness' : instrument, 'liveness' : live, 'valence' : valence, 'tempo' : tempo,
                            'duration_ms' : duration, 'time_signature' : time})
print(spotify_df.shape)
spotify_df.head()
```
